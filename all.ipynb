```python
import numpy as np
import cv2
import joblib
import Sliding as sd
from imutils.object_detection import non_max_suppression
from skimage.feature import hog
from skimage import color
from skimage.transform import pyramid_gaussian
import matplotlib.pyplot as plt
import os
from PIL import Image

# Load model
model = joblib.load('models/model_crop.dat')

# Path to the input folder
input_folder = 'drone new/test_new'

# Path to the output folder
output_folder = 'bahan_ppt'

# Create output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Loop through each file in the input folder
for filename in os.listdir(input_folder):
    # Read image
    image = cv2.imread(os.path.join(input_folder, filename))
    image = cv2.resize(image, (400, 256))
    # size = (128, 64)
    size = (64, 128)
    step_size = (9, 9)
    downscale = 1.25

    # List to store detections
    detections = []

    # Current scale of the image
    scale = 0

    # Loop through Gaussian pyramid
    for im_scaled in pyramid_gaussian(image, downscale=downscale):
        # If the resized image is smaller than the minimum desired size
        if im_scaled.shape[0] < size[1] or im_scaled.shape[1] < size[0]:
            break

        # Loop through sliding window
        for (x, y, window) in sd.sliding_window(im_scaled, size, step_size):
            if window.shape[0] != size[1] or window.shape[1] != size[0]:
                continue
            window = color.rgb2gray(window)

            # Extract HOG features
            fd = hog(window, orientations=9, pixels_per_cell=(8, 8), visualize=False, cells_per_block=(3, 3))
            fd = fd.reshape(1, -1)

            # Predict using the model
            pred = model.predict(fd)
            if pred == 1:
                if model.decision_function(fd) > 0.5:
                    detections.append((int(x * (downscale ** scale)), int(y * (downscale ** scale)),
                                       model.decision_function(fd),
                                       int(size[0] * (downscale ** scale)),
                                       int(size[1] * (downscale ** scale))))

        scale += 1

    # Array of rectangles
    rects = np.array([[x, y, x + w, y + h, cfd[0]] for (x, y, cfd, w, h) in detections])
    sc = [score[0] for (x, y, score, w, h) in detections]

    # Non-maxima suppression
    sc = np.array(sc)
    pick = non_max_suppression(rects[:, :4], probs=sc, overlapThresh=0.01)

    # Print sorted scores
    print("sc: ", sorted(sc))

    # Read original image for EXIF data
    img_pil = Image.open(os.path.join(input_folder, filename))
    exif_data = img_pil.info.get('exif')

    # Loop through all picked bounding boxes
    for i, (x1, y1, x2, y2) in enumerate(pick):
        # Crop image
        crop_img = image[y1:y2, x1:x2]

        # Print score and show the cropped image
        print(f"Cropped Image {i} Score: {sc[i]}")
        plt.imshow(cv2.cvtColor(crop_img, cv2.COLOR_BGR2RGB))
        plt.show()

        # Save cropped image only if score is higher than threshold
        if sc[i] > 0.1:
            # Convert color from BGR to RGB before saving with Matplotlib or PIL
            crop_img_rgb = cv2.cvtColor(crop_img, cv2.COLOR_BGR2RGB)

            # Save using Matplotlib
            plt.imshow(crop_img_rgb)
            plt.axis('off')
            output_filename = os.path.join(output_folder, f'crop_{i}_{filename}')
            plt.savefig(output_filename)
            plt.close()

            # Save cropped image with EXIF metadata
            crop_img_pil = Image.fromarray(crop_img_rgb)
            if exif_data:
                crop_img_pil.save(output_filename, exif=exif_data)
            else:
                crop_img_pil.save(output_filename)

    # Display EXIF data
    exif_data = img_pil._getexif()
    print(f"EXIF Data for {filename}:")
    if exif_data is not None and 0x8825 in exif_data:  # 0x8825 is the code for GPSInfo
        gps_info = exif_data[0x8825]
        lat = gps_info[2][0] + gps_info[2][1] / 60 + gps_info[2][2] / 3600
        if gps_info[1] == 'S':
            lat = -lat

        lon = gps_info[4][0] + gps_info[4][1] / 60 + gps_info[4][2] / 3600
        if gps_info[3] == 'W':
            lon = -lon

        print(f"Latitude: {lat}, Longitude: {lon}")
    else:
        print("No GPS Info")
```

```python
import numpy as np 
import cv2
import joblib
import Sliding as sd
from imutils.object_detection import non_max_suppression
from skimage.feature import hog
from skimage import color, data, exposure
from skimage.transform import pyramid_gaussian
import matplotlib.pyplot as plt
import glob
import os
from PIL import Image

# Load model
model = joblib.load('models/model_crop.dat')

# Path ke folder input
input_folder = 'drone new/test'

# Path ke folder output
output_folder = 'bahan_ppt'

# Membuat folder output jika belum ada
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Loop melalui setiap file dalam folder input
for filename in os.listdir(input_folder):
    # Baca gambar
    image = cv2.imread(os.path.join(input_folder, filename))
    image = cv2.resize(image, (400, 256))
    size = (64, 128)
    # size = (128, 64)
    step_size = (9, 9)
    downscale = 1.25

    # List untuk menyimpan deteksi
    detections = []

    # Skala saat ini dari gambar
    scale = 0

    # Loop melalui piramida Gaussian
    for im_scaled in pyramid_gaussian(image, downscale=downscale):
        # Jika gambar yang diperkecil lebih kecil dari ukuran minimum yang diinginkan
        if im_scaled.shape[0] < size[1] or im_scaled.shape[1] < size[0]:
            break

        # Loop melalui sliding window
        for (x, y, window) in sd.sliding_window(im_scaled, size, step_size):
            if window.shape[0] != size[1] or window.shape[1] != size[0]:
                continue
            window = color.rgb2gray(window)

            # Ekstraksi fitur HOG
            fd = hog(window, orientations=9, pixels_per_cell=(8, 8), visualize=False, cells_per_block=(3, 3))
            fd = fd.reshape(1, -1)

            # Prediksi menggunakan model
            pred = model.predict(fd)
            if pred == 1:
                if model.decision_function(fd) > 0.5:
                    detections.append((int(x * (downscale ** scale)), int(y * (downscale ** scale)),
                                       model.decision_function(fd),
                                       int(size[0] * (downscale ** scale)),
                                       int(size[1] * (downscale ** scale))))

        scale += 1

    # Array of rectangles
    rects = np.array([[x, y, x + w, y + h, cfd[0]] for (x, y, cfd, w, h) in detections])
    sc = [score[0] for (x, y, score, w, h) in detections]

    # Non-maxima suppression
    sc = np.array(sc)
    pick = non_max_suppression(rects, probs=sc, overlapThresh=0.01)

    # Cetak skor terurut
    print("sc: ", sorted(sc))

    # Mencari indeks bounding box dengan nilai sc tertinggi
    idx = np.argmax(sc)

    # Mendapatkan koordinat bounding box dengan nilai sc tertinggi
    (x, y, _, w, h) = detections[idx]

    # Mengambil bounding box dengan nilai sc tertinggi dari gambar asli
    crop_img = image[y:y + h, x:x + w]

    # Menyimpan gambar yang di-crop hanya jika nilai sc lebih dari 2
    if sc[idx] > 2:
        plt.imshow(crop_img)
        plt.axis('on')  # Matikan sumbu

        # Simpan gambar dengan nama unik dalam folder output
        output_filename = os.path.join(output_folder, f'crop_{filename}')
        plt.savefig(output_filename)
        plt.close()

        # Membaca data EXIF dari gambar asli
        img_pil = Image.open(os.path.join(input_folder, filename))
        exif_data = img_pil._getexif()

        # Simpan gambar crop tanpa kehilangan metadata EXIF
        crop_img_pil = Image.fromarray(crop_img)
        if exif_data:
            crop_img_pil.save(output_filename, exif=img_pil.info['exif'])
        else:
            crop_img_pil.save(output_filename)

    # Membaca data EXIF dari gambar asli
    img_pil = Image.open(os.path.join(input_folder, filename))
    exif_data = img_pil._getexif()

    # Menampilkan data EXIF
    print(f"EXIF Data for {filename}:")
    if exif_data is not None and 0x8825 in exif_data:  # 0x8825 adalah kode untuk GPSInfo
        gps_info = exif_data[0x8825]
        lat = gps_info[2][0] + gps_info[2][1] / 60 + gps_info[2][2] / 3600
        if gps_info[1] == 'S':
            lat = -lat

        lon = gps_info[4][0] + gps_info[4][1] / 60 + gps_info[4][2] / 3600
        if gps_info[3] == 'W':
            lon = -lon

        print(f"Latitude: {lat}, Longitude: {lon}")
    else:
        print("No GPS Info")
```

```python
import numpy as np
import cv2
import joblib
import Sliding as sd
from imutils.object_detection import non_max_suppression
from skimage.feature import hog
from skimage import color
from skimage.transform import pyramid_gaussian
import matplotlib.pyplot as plt
import os
from PIL import Image

# Load model
model = joblib.load('models/model_crop.dat')

# Path ke folder input
input_folder = 'drone new/test_new'

# Path ke folder output
output_folder = 'bahan_ppt'

# Membuat folder output jika belum ada
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Loop melalui setiap file dalam folder input
for filename in os.listdir(input_folder):
    # Baca gambar
    image = cv2.imread(os.path.join(input_folder, filename))
    image = cv2.resize(image, (400, 256))
    size = (64, 128)
    step_size = (9, 9)
    downscale = 1.25

    # List untuk menyimpan deteksi
    detections = []

    # Skala saat ini dari gambar
    scale = 0

    # Loop melalui piramida Gaussian
    for im_scaled in pyramid_gaussian(image, downscale=downscale):
        # Jika gambar yang diperkecil lebih kecil dari ukuran minimum yang diinginkan
        if im_scaled.shape[0] < size[1] or im_scaled.shape[1] < size[0]:
            break

        # Loop melalui sliding window
        for (x, y, window) in sd.sliding_window(im_scaled, size, step_size):
            if window.shape[0] != size[1] or window.shape[1] != size[0]:
                continue
            window = color.rgb2gray(window)

            # Ekstraksi fitur HOG
            fd = hog(window, orientations=9, pixels_per_cell=(8, 8), visualize=False, cells_per_block=(3, 3))
            fd = fd.reshape(1, -1)

            # Prediksi menggunakan model
            pred = model.predict(fd)
            if pred == 1:
                if model.decision_function(fd) > 0.5:
                    detections.append((int(x * (downscale ** scale)), int(y * (downscale ** scale)),
                                       model.decision_function(fd),
                                       int(size[0] * (downscale ** scale)),
                                       int(size[1] * (downscale ** scale))))

        scale += 1

    if detections:
        # Array of rectangles
        rects = np.array([[x, y, x + w, y + h, cfd[0]] for (x, y, cfd, w, h) in detections])
        sc = [score[0] for (x, y, score, w, h) in detections]

        # Non-maxima suppression
        sc = np.array(sc)
        pick = non_max_suppression(rects, probs=sc, overlapThresh=0.01)

        # Cetak skor terurut
        print("sc: ", sorted(sc))

        # Mencari indeks bounding box dengan nilai sc tertinggi
        idx = np.argmax(sc)

        # Mendapatkan koordinat bounding box dengan nilai sc tertinggi
        (x, y, _, w, h) = detections[idx]

        # Mengambil bounding box dengan nilai sc tertinggi dari gambar asli
        crop_img = image[y:y + h, x:x + w]

        # Menyimpan gambar yang di-crop hanya jika nilai sc lebih dari 2
        if sc[idx] > 2:
            # Konversi warna dari BGR ke RGB sebelum menyimpan dengan Matplotlib atau PIL
            crop_img_rgb = cv2.cvtColor(crop_img, cv2.COLOR_BGR2RGB)

            # Simpan menggunakan Matplotlib
            plt.imshow(crop_img_rgb)
            plt.axis('on')
            output_filename = os.path.join(output_folder, f'crop_{filename}')
            plt.savefig(output_filename)
            plt.close()

            # Membaca data EXIF dari gambar asli
            img_pil = Image.open(os.path.join(input_folder, filename))
            exif_data = img_pil.info.get('exif')

            # Simpan gambar crop dengan metadata EXIF
            crop_img_pil = Image.fromarray(crop_img_rgb)
            if exif_data:
                crop_img_pil.save(output_filename, exif=exif_data)
            else:
                crop_img_pil.save(output_filename)

        # Membaca data EXIF dari gambar asli
        img_pil = Image.open(os.path.join(input_folder, filename))
        exif_data = img_pil._getexif()

        # Menampilkan data EXIF
        print(f"EXIF Data for {filename}:")
        if exif_data is not None and 0x8825 in exif_data:  # 0x8825 adalah kode untuk GPSInfo
            gps_info = exif_data[0x8825]
            lat = gps_info[2][0] + gps_info[2][1] / 60 + gps_info[2][2] / 3600
            if gps_info[1] == 'S':
                lat = -lat

            lon = gps_info[4][0] + gps_info[4][1] / 60 + gps_info[4][2] / 3600
            if gps_info[3] == 'W':
                lon = -lon

            print(f"Latitude: {lat}, Longitude: {lon}")
        else:
            print("No GPS Info")
    else:
        print(f"No detections for {filename}")
```
